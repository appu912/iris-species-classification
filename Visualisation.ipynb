```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

iris = pd.read_csv('Iris.csv')
iris.head()
```

```python
iris.shape
```

```python
iris = iris[['SepalLengthCm' , 'SepalWidthCm' , 'PetalLengthCm' , 'PetalWidthCm' , 'Species']]
print(iris.columns)
```

```python
# How many datapoints for each class(species) are present?
iris["Species"].value_counts()
```

```python
sns.set_style('darkgrid')
graph = sns.FacetGrid(iris , hue = 'Species' , height = 8)
graph = graph.map(plt.scatter , 'SepalLengthCm' , 'SepalWidthCm')
graph.add_legend()

plt.show()
```

### Observation
Using sepal length and sepal width we can distinguish between blue and other points by drawing a line. Thus setosa can be distinguished using sepal length and sepal width. Thus setosa is linearly seperable.

We can plot for all combinations of length and then check.

```python
plt.close()
sns.set_style('whitegrid')
sns.pairplot(iris , hue = 'Species' , height = 5)
plt.show()

# 4C2 = 6 pairs of features 
# We can omit the last row as the have just interchange x and y.
# Diagonals are for PDFs
```

### Observations

1. Petal length and petal width are the most important features that can be used for distinguishing as the data is more distinguisable their.
2. While Setosa can be easily identified(linearly seperable) , Virginica and versicolor have some overlap(almost linearly seperable)
3. We can find 'lines' and simple 'if - else' conditions to build a model to classify the flower types.

## Univariate Analysis

```python
# Anaysing petal length
graph = sns.FacetGrid(iris , hue = 'Species' , height = 5)
graph.map(sns.distplot , 'PetalLengthCm')
graph.add_legend()
plt.show()
```

```python
# Analysing petal width
graph = sns.FacetGrid(iris , hue = 'Species' , height = 5)
graph.map(sns.distplot , 'PetalWidthCm')
graph.add_legend()
plt.show()
```

```python
# analysing Sepal Length
graph = sns.FacetGrid(iris , hue = 'Species' , height = 5)
graph.map(sns.distplot , 'SepalLengthCm')
graph.add_legend()
plt.show()
```

```python
# analysing Sepal Width
graph = sns.FacetGrid(iris , hue = 'Species' , height = 5)
graph.map(sns.distplot , 'SepalWidthCm')
graph.add_legend()
plt.show()
```

### Observations

**Petal Length**

Setosa is well seperated from other species. Their is little bit of overlap between virginica and versicolor.

**Petal Width**

Setosa is slightly overlapped with other species. Their is little bit of overlap between virginica and versicolor.

**Sepal Length**

All three species are highly overlapped.

**Sepal Width**

All three species are highly overlapped even more than sepal length.

Thus we order the importance of features as:

    petal_length > petal_width >>> sepal_length > sepal_width
